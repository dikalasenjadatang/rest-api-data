```python
import json
import os
import pandas as pd
import numpy as np
from pymongo import MongoClient
from datetime import datetime
```

```python
MONGO_URI = 'mongodb://localhost:27017/'  # Update with your MongoDB URI
DATABASE_NAME = 'E2E_TESTING'
DATA_COLLECTION_NAME = 'data_scrape'
TOPIK = 'KEMENTRIAN_Z'
```

```python
# Create MongoDB client
client_db = MongoClient(MONGO_URI)
db = client_db[DATABASE_NAME]
topik_collection = db[TOPIK]
```

```python
document = topik_collection.find_one({'_id': 'news'})

data = document['data'] if document else []
# Convert to DataFrame
df = pd.DataFrame(data)
df.head()
```

```python
substring = 'DPR'

# Filter berdasarkan dua kolom
filtered_df = df[df['judul'].str.contains(substring, case=False) | df['ringkasan'].str.contains(substring, case=False)]

print(len(filtered_df))
```
